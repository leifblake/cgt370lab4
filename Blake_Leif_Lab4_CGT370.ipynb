```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)

# Generate random sales data for 10 days for 3 products
sales_data = np.random.randint(1, 20, size=(10, 3))  # 10 rows, 3 columns

# Create a date range for 10 consecutive days
dates = pd.date_range(start='2025-02-01', periods=10, freq='D')

# Create the DataFrame
sales_df = pd.DataFrame(sales_data, columns=["Product 1", "Product 2", "Product 3"], index=dates)

sales_df
```

```python
price_array = np.array([1, 2, 3])
```

```python
print("Shape of sales data:", sales_df.shape)
print("Shape of price array:", price_array.shape)
```

```python
# Calculate total sales for each product using np.dot
total_sales = np.dot(sales_df, price_array)

total_sales
```

```python
# Add total sales as a new column to the DataFrame
sales_df['Total Sales'] = total_sales

sales_df
```
